```python
import pandas as pd
```

```python
#original file
review_df = pd.read_csv("train.csv", index_col=0)
review_df.rename(columns = {'rating': 'label'}, inplace = True)
review_df = review_df.sample(frac=1)
review_df.head()
```

```python
review_df.describe()
```

```python
review_df.groupby('label').describe()
```

```python
review_df['length'] = review_df['review'].apply(len)
review_df.head()
```

```python
import string

review_df['Word Count'] = [len(review.split()) for review in review_df['review']]

review_df['Uppercase Char Count'] = [sum(char.isupper() for char in review) \
                              for review in review_df['review']]                           

review_df['Special Char Count'] = [sum(char in string.punctuation for char in review) \
                            for review in review_df['review']] 

review_df.head()
```
